## 1. Simple List Adder

Let's create a function that adds numbers in a list:

```python
# Simple list adder function
def add_list(numbers):
    """Add all numbers in a list"""
    return sum(numbers)

# Test it
my_numbers = [5, 10, 15, 20, 25]
result = add_list(my_numbers)

print(f"Numbers: {my_numbers}")
print(f"Sum: {result}")
```

## 2. Working with Lists

Jupyter notebooks are great for exploring data step by step:

```python
# Create a list of fruits
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']
print("My fruits:", fruits)
```

```python
# Add more fruits
fruits.append('fig')
fruits.append('grape')
print("Updated list:", fruits)
print("Total fruits:", len(fruits))
```

## 3. Simple Calculations

You can use Jupyter like a calculator:

```python
a = 42
b = 17

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} × {b} = {a * b}")
print(f"{a} ÷ {b} = {a / b:.2f}")
```

## 4. Data Visualization

Jupyter notebooks can display charts inline:

```python
import matplotlib.pyplot as plt

# Create sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='steelblue')
plt.title('Simple Bar Chart Example', fontsize=16)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Line plot
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, linewidth=2, color='green')
plt.title('Sine Wave', fontsize=16)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True, alpha=0.3)
plt.show()
```

## 5. Working with DataFrames

Pandas DataFrames display nicely in Jupyter:

```python
import pandas as pd

# Create a simple dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, 30, 35, 28, 32],
    'Score': [85, 92, 78, 95, 88],
    'Grade': ['B', 'A', 'C', 'A', 'B']
}

df = pd.DataFrame(data)
df
```

```python
# Calculate some statistics
print("Average age:", df['Age'].mean())
print("Average score:", df['Score'].mean())
print("\nScore statistics:")
print(df['Score'].describe())
```

The beauty of Jupyter is you can run cells in any order and experiment:
